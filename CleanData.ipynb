```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

# Cleans and encodes customer data for machine learning.

```python
# Load the dataset
df = pd.read_csv("./Data/Telco-Customer-Churn.csv")
df
```

```python
df.columns
```

```python
print(df.head())
print(df.info())
```

# Replace spaces in 'TotalCharges' with NaN and convert to numeric

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

# Drop rows with missing values in 'TotalCharges'

```python
df.dropna(subset=['TotalCharges'], inplace=True)
```

# Make a copy to avoid SettingWithCopyWarning

```python
df_cleaned = df.copy()
```

# Identify categorical and numerical columns

```python
categorical_cols = df_cleaned.select_dtypes(include='object').columns.tolist()
numerical_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

# Separate binary and multi-class categorical columns

```python
binary_cols = [col for col in categorical_cols if df_cleaned[col].nunique() <= 2]
multi_class_cols = [col for col in categorical_cols if df_cleaned[col].nunique() > 2]
```

# Apply Label Encoding for binary categorical columns

```python
le = LabelEncoder()
for col in binary_cols:
    df_cleaned[col] = le.fit_transform(df_cleaned[col])
```

# Apply One-Hot Encoding for multi-class categorical columns

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=multi_class_cols, drop_first=True)
```

```python
print(df_cleaned.head())
print(df_cleaned.info())
```

# You can save the cleaned DataFrame to a new CSV file

```python
df_cleaned.to_csv('./Data/cleanedChurn.csv', index=False)
```
